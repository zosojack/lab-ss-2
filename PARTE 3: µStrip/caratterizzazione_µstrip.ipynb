plot S11 [in dB] su frequenza [in MHz]

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('data/vector_analyzer_strip.csv', skiprows=1)
freq = data[:, 0]  # in MHz
S11 = data[:, 1] #/ np.max(np.abs(data[:, 1]))  # Normalized S11

fig, ax = plt.subplots(figsize=(8, 5), dpi=200)

# Curva 1: misura, asse Y sinistro
ln1 = ax.plot(freq, S11, color='blue', label='Misura', linewidth=1.0)
ax.set_xlabel('Frequenza [MHz]')
ax.set_ylabel('S11 misura [dB]', color='blue')
ax.tick_params(axis='y', labelcolor='blue')
#ax.grid(True, alpha=0.3)

# Range richiesto 0  -3 (asse invertito)
ax.set_ylim(-4, 0)

# Curva 2: simulazione, asse Y destro
ax2 = ax.twinx()
ln2 = ax2.plot(x_sim, y_sim, color='red', linestyle='-', label='Simulazione', linewidth=1.0)
ax2.set_ylabel('S11 simulazione [dB]', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Range richiesto -10  -30 (asse invertito)
ax2.set_ylim(-35, -8)

# Legenda combinata
lines = ln1 + ln2
labels = [l.get_label() for l in lines]
ax.legend(lines, labels, loc='best')

fig.tight_layout()
#fig.savefig("S11-vs-frequenza.pdf", bbox_inches='tight')
plt.show()
```
