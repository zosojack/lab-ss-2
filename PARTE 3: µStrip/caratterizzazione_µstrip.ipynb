# MICROSTRIP

plot S11 [in dB] su frequenza [in MHz]

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('data/vector_analyzer_strip.csv', skiprows=1)
freq = data[:, 0]  # in MHz
S11 = data[:, 1]

plt.figure(figsize=(8, 5), dpi=200)
plt.plot(freq, S11, color='black')
plt.xlabel('Frequenza [MHz]')
plt.ylabel('S11 [dB]')
#plt.title('S11 vs Frequenza')
#plt.grid()
plt.tight_layout()
plt.show()
plt.savefig("S11-vs-frequenza.pdf")
```
