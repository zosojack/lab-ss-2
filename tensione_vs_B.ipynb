```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from CSV file
data = np.loadtxt('data/tensione_vs_b.csv', skiprows=1)  # Adjust delimiter and skiprows as needed
B = data[:, 0]  # Assuming "B" is in the first column
err_B = data[:, 1]
tensione = data[:, 2]  # Assuming "tensione" is in the second column
err_tensione = data[:, 3]

# Plot the data
plt.figure(figsize=(8, 6))
plt.figure(figsize=(8, 6))
plt.errorbar(
    B, tensione, xerr=err_B, yerr=err_tensione,
    fmt='o',  # Solo marker, senza linee tra i punti
    markersize=0.5,  # Dimensione dei marker leggermente aumentata
    elinewidth=1,  # Spessore delle barre d'errore ridotto per leggibilit√†
    capsize=0.8,  # Aggiunge terminali alle barre d'errore
    capthick=1,  # Spessore dei terminali
    color='black',  # Colore neutro per evitare distrazioni
    label='Dati'
)
plt.errorbar(B, tensione, xerr=err_B, yerr=err_tensione, fmt='o', markersize=1, linestyle='-', color='b', label='Tensione vs B')
plt.xlabel('B')
plt.ylabel('Tensione')
plt.title('Tensione vs B con barre di errore')
plt.legend()
plt.grid(linestyle='--')
plt.show()
```
