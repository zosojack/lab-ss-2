# PRE SPECCHIO

```python
import numpy as np
import matplotlib.pyplot as plt

# Legge il file ignorando la prima riga (intestazione)
dati = np.loadtxt("tensione_potenza_pre.txt", skiprows=1)

# Estrae le colonne
potenza = dati[:, 0]   # Prima colonna: Potenza [mW]
tensione = np.abs(dati[:, 2])  # Terza colonna: Tensione [mV]

# Plot
plt.figure(figsize=(8, 5), dpi=200)
plt.plot(tensione, potenza, marker='o', linestyle='-', markersize=3)
plt.ylabel("Potenza [mW]")
plt.xlabel("Tensione [mV]")
plt.title("Potenza vs Tensione | PRO specchio")
#plt.grid(linestyle='--')
plt.show()
```

# POST SPECCHIO

```python
# Legge il file ignorando la prima riga (intestazione)
dati_post = np.loadtxt("tensione_potenza_post.txt", skiprows=1)

# Estrae le colonne
potenza_post = dati_post[:, 0]   # Prima colonna: Potenza [mW]
tensione_post = np.abs(dati_post[:, 2])  # Terza colonna: Tensione [mV]

# Plot
plt.figure(figsize=(6, 4), dpi=200)
plt.plot(tensione_post, potenza_post, marker='o', linestyle='-', markersize=3)
plt.ylabel("Potenza [mW]")
plt.xlabel("Tensione [mV]")
plt.title("Potenza vs Tensione | POST specchio")
#plt.grid(linestyle='--')
plt.show()
```

# RAPPORTO tra le tensioni

Questo dovrebbe significare che il 96% del fascio passa, perché qualcosa attorno al 4 viene trasmesso. Nota bene: noi siamo interessati a quello riflesso non a quello trasm.

```python
rapporto = tensione_post/tensione * 100

print(rapporto)
```

```python
# Plot
plt.figure(figsize=(6, 4), dpi=200)
plt.plot(potenza_post[4:], rapporto[4:], marker='o', linestyle='-', markersize=3)
plt.xlabel("Potenza [mW]")
plt.ylabel("% Tensione trasmessa")
plt.title(r"Potenza vs Rapporto Tensioni | PRE/POST$\times$100")
#plt.grid(linestyle='--')
plt.show()
```
