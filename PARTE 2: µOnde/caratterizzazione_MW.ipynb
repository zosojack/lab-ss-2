# GENERATORE MICROONDE: SEGNALE PRE E POST AMPLI

---

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data/phase_noise_mw.csv', skiprows=1)
freq = data[:, 0] * 1_000_000  # Convert frequency to Hz
phase_noise = - data[:, 1]
err_phase_noise = data[:, 2]

plt.figure(figsize=(10, 6))
plt.errorbar(freq, phase_noise, yerr=err_phase_noise, color='blue')
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Noise (dBc/Hz)')
plt.title('Generatore microonde')
plt.tight_layout()
plt.show()
```

# AMPLIFICATORE

```python
data = np.loadtxt('data/phase_noise_ampli.csv', skiprows=1)
freq = data[:, 0] * 1_000_000  # Convert frequency to Hz
phase_noise = data[:, 1]
err_phase_noise = data[:, 2]

plt.figure(figsize=(10, 6))
plt.errorbar(freq, phase_noise, yerr=err_phase_noise, color='blue')
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Noise (dBc/Hz)')
plt.title('Segnale amplificato')
plt.tight_layout()
plt.show()
```
