# GENERATORE MICROONDE: SEGNALE PRE E POST AMPLI

---

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data/phase_noise_mw.csv', skiprows=1)
freq = data[:, 0] * 1_000_000  # Convert frequency to Hz
phase_noise = - data[:, 1]
err_phase_noise = data[:, 2]

plt.figure(figsize=(8, 5), dpi=200)
plt.errorbar(freq, phase_noise, yerr=err_phase_noise, color='black', label='Generatore MW')
plt.xscale('log')
plt.xlabel('Distanza dalla portante (2.87 GHz) [Hz]')
plt.ylabel('Rumore di fase [dBc/Hz]')
#plt.title('Segnale amplificato')
plt.tight_layout()
#plt.legend()
plt.show()
```

---

# AMPLIFICATORE

```python
data = np.loadtxt('data/phase_noise_ampli.csv', skiprows=1)
freq2 = data[:, 0] * 1_000_000  # Convert frequency to Hz
phase_noise2 = data[:, 1]
err_phase_noise2 = data[:, 2]

plt.figure(figsize=(8, 5), dpi=200)
plt.errorbar(freq2, phase_noise2, yerr=err_phase_noise2, color='black')
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase Noise (dBc/Hz)')
#plt.title('Segnale amplificato')
plt.tight_layout()
plt.show()
#plt.savefig("phase-noise-vs-frequenza.pdf")
```

# INSIEME

```python
plt.figure(figsize=(8, 5), dpi=200)
plt.errorbar(freq, phase_noise, yerr=err_phase_noise, color='blue', label='Generatore MW')
plt.errorbar(freq2, phase_noise2, yerr=err_phase_noise2, color='red', label='Amplificatore')
plt.xscale('log')
plt.xlabel('Distanza dalla portante (2.87 GHz) [Hz]')
plt.ylabel('Rumore di fase [dBc/Hz]')
#plt.title('Segnale amplificato')
plt.tight_layout()
plt.legend()
plt.show()
```
