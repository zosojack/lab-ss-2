```python
import numpy as np
import matplotlib.pyplot as plt

# Legge il file ignorando la prima riga (intestazione)
dati = np.loadtxt("data/tensione_potenza_pre.csv", skiprows=1)

# Estrae le colonne
potenza = dati[:, 0]   # Prima colonna: Potenza [mW]
tensione = np.abs(dati[:, 2])  # Terza colonna: Tensione [mV]
err_tens = dati[:, 4]

# Plot
plt.figure(figsize=(8, 5), dpi=200)
plt.plot(tensione, potenza, marker='o', linestyle='-', markersize=3)
plt.ylabel("Potenza [mW]")
plt.xlabel("Tensione [mV]")
plt.title("Potenza vs Tensione | PRE specchio")
#plt.grid(linestyle='--')
plt.show()
```

### RAPPORTO SEGNALE RUMORE PRE

```python
snr_pre = tensione / err_tens

# Plot
plt.figure(figsize=(6, 4), dpi=200)
plt.plot(tensione[1:], snr_pre[1:], marker='o', linestyle='-', markersize=3)
plt.xlabel("Tensione [mV]")
plt.ylabel("% Tensione trasmessa")
plt.title(r"SNR su Segnale | PRE")
#plt.grid(linestyle='--')
plt.show()
```
