# Andamento Segnale

```python
def function(x, A, P):
    return A/(1+P/x)
```

### Leggo

```python
import numpy as np

data = np.loadtxt("data/snr_fotoluminescenza.csv", skiprows=1)
potenza = data[:,0] # potenza (ma in corrente)
segnale = data[:,1]
rumore  = data[:,2]
SNR = data[:,3]
```

### Plotto

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5), dpi=200)
plt.plot(potenza, segnale, marker='x', label="Segnale", linestyle='-', markersize=3, color='blue')
plt.plot(potenza, rumore, marker='o', label="Rumore", linestyle='-', markersize=3, color='green')
plt.plot(potenza, SNR, marker='^', label="SNR", linestyle='-', markersize=3, color='red')
plt.xlabel("Corrente in entrata laser [mA]")
#plt.xticks(x2_ticks, x2_ticks)
plt.ylabel("[mV]")
plt.title("Segnale, Rumore e SNR su Corrente")
#plt.grid(linestyle='--')
plt.legend()
plt.show()
```

## FIT SEGNALE

```python
from scipy.optimize import curve_fit
```

```python
plt.figure(figsize=(8, 5), dpi=200)
plt.plot(potenza, segnale, marker='x', label="Segnale", linestyle='-', markersize=3, color='blue')

## fit ##
popt, pcov = curve_fit(function, potenza, segnale)
x = np.linspace(175, 390, 250)
y_fit = function(x, *popt)
plt.plot(x, y_fit, label="Fit", linestyle='--', color='blue')

plt.xlabel("Corrente in entrata laser [mA]")
#plt.xticks(x2_ticks, x2_ticks)
plt.ylabel("[mV]")
plt.title("Segnale, Rumore e SNR su Corrente")
#plt.grid(linestyle='--')
plt.legend()
plt.show()
```
